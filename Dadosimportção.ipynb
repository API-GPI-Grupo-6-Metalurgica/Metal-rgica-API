```python
imp1 = pd.read_csv(arquivo_1, low_memory=False, sep=',', encoding='UTF-8', quotechar='"')
imp2 = pd.read_csv(arquivo_2, low_memory=False, sep=',', encoding='UTF-8', quotechar='"')
imp3 = pd.read_csv(arquivo_3, low_memory=False, sep=',', encoding='UTF-8', quotechar='"')
```
