```python
origem = '/content/drive/MyDrive/Data/COMEXSTAT/'
```

```python
arquivo_1 = origem + 'EXP_2021_MUN.csv'
arquivo_2 = origem + 'EXP_2022_MUN.csv'
ncm = origem + 'NCM.csv'
pais = origem + 'PAIS.csv'
uf_mun =origem + 'UF_MUN.csv'
ncm_sh = origem + 'NCM_SH.csv'
```

```python
exp21 = pd.read_csv(arquivo_1,low_memory=False,sep=';',encoding='UTF-8')
exp22 = pd.read_csv(arquivo_2,low_memory=False,sep=';',encoding='UTF-8')
expncm = pd.read_csv(ncm,low_memory=False,sep=';',encoding='latin1')
exppais = pd.read_csv(pais,low_memory=False,sep=';',encoding='latin1')
expufmun = pd.read_csv(uf_mun,low_memory=False,sep=';',encoding='latin1')
expsh = pd.read_csv(ncm_sh,low_memory=False,sep=';',encoding='latin1')
```

```python
exp_final=pd.concat([exp21,exp22], ignore_index=True)
exp_final.info()
```

```python
exp_final = exp_final.merge(expsh[['CO_SH4','NO_SH4_POR']],left_on='SH4', right_on='CO_SH4',how='left')
```

```python
exp_final = exp_final.merge(exppais[['CO_PAIS','NO_PAIS']],on='CO_PAIS',how='left')
```

```python
exp_final = exp_final.merge(expufmun[['CO_MUN_GEO','NO_MUN_MIN']],left_on='CO_MUN', right_on='CO_MUN_GEO',how='left')
```
